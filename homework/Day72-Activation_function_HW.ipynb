# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
```

```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

#x=np.arange(0,1.0,0.01)
x = plt.linspace(-5,5,100)

#resize the X and Y axes
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
#plt.gca().yaxis.set_major_locator(plt.MultipleLocator(1))

ReLUList = []
for i in range(len(x)):
    ReLUList.append(max(0, x[i]))

#print(ReLU(x))
plt.plot(x, ReLUList, 'r')
plt.show()
```

```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

#x=np.arange(0,1.0,0.01)
x = plt.linspace(-5,5,100)

#resize the X and Y axes
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
#plt.gca().yaxis.set_major_locator(plt.MultipleLocator(1))

dReLUList = []
for i in range(len(x)):
    if x[i] > 0:
        dReLUList.append(1)
    else:
        dReLUList.append(0)

#print(dReLU(x))
plt.plot(x, dReLUList, 'r')
plt.show()
```
